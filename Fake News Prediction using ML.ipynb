**Fake News Prediction**

```python
**Dataset: The data that will be used was gathered from Twitter around the time of the 2016 US General Election**
**Title- the title of the news tweet**
**Text- the actual text of the news tweet**
**Label- the label of the news tweet, either Fake or Real**
```

```python
import numpy as np
import pandas as pd
import itertools
```

```python
df=pd.read_csv("news.csv")
df
```

```python
df=df.drop(columns=['Unnamed: 0'])
df
```

```python
print(df['label'])
```

```python
x=df['text']
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer=TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
tfidf_train=tfidf_vectorizer.fit_transform(x_train) 
tfidf_test=tfidf_vectorizer.transform(x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(tfidf_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
y_pred=model.predict(tfidf_test)
score=accuracy_score(y_test,y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pa=PassiveAggressiveClassifier(max_iter=50)
pa.fit(tfidf_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
y_pred=pa.predict(tfidf_test)
score=accuracy_score(y_test,y_pred)
print(f'Accuracy: {round(score*100,2)}%')
```

### Comparison of Naive Bayes and Passive Aggressive Classifier
- **Model Performance:** The Passive Aggressive Classifier outperforms Naive Bayes by capturing feature interactions and dependencies.
- **Feature Independence Assumption:** Naive Bayes assumes feature independence, leading to suboptimal performance with correlated features.
- **Complex Relationships:** The Passive Aggressive Classifier effectively models complex relationships, enhancing accuracy in real-world applications.
- **Robustness to Noise:** It is more resilient to noise and irrelevant features, improving learning from data.
- **Adaptability:** The Passive Aggressive Classifier adapts better to changes in data distribution, making it suitable for dynamic datasets.

### Conclusion
The Passive Aggressive Classifier is preferred for its ability to capture feature dependencies, resulting in improved accuracy and robustness. This will be documented in my GitHub repository to emphasize its advantages over Naive Bayes.
